```python
import pandas as pd
import numpy as np
df = pd.read_csv('산업분류.csv',encoding='cp949')
df = df.drop(columns = ['시도명','시군구명','법정동명','행정동코드','법정동코드','지번주소','도로명주소','신우편번호','경도','위도'])
df=df.sort_values(by=['행정동명'], axis=0)
df
```

```python
count = pd.DataFrame(df.groupby(['행정동명','상권업종대분류명']).size())
count
```

```python
count.reset_index(drop=False, inplace=True)
count.rename(columns={0:'점포 수'}, inplace=True)
count
```

```python
#각 지역마다 업종별 카운트 
count=count.sort_values(by=['행정동명','상권업종대분류명'])
count
```

```python
count = count.sort_values([ '상권업종대분류명', '행정동명'])
count
```

```python
activity_count=count[count.상권업종대분류명=='문화레저']
activity_count
```

```python
food_count=count[count.상권업종대분류명=='음식']
del food_count['행정동명']
del food_count['상권업종대분류명']
food_count=food_count.reset_index(drop=True)
food_count
```

```python
some_count=count[count.상권업종대분류명=='소매']
del some_count['행정동명']
del some_count['상권업종대분류명']
some_count=some_count.reset_index(drop=True)
some_count
```

```python
sleep_count=count[count.상권업종대분류명=='숙박']
del sleep_count['행정동명']
del sleep_count['상권업종대분류명']
sleep_count=sleep_count.reset_index(drop=True)
sleep_count
```

```python
jejudata19 = pd.read_csv('jejudata19.csv',encoding='cp949')
jejudata19 = jejudata19.drop(columns = ['년월', '시도명', '총 유동인구','데이터기준일자'])
jejudata19
```

```python
food=jejudata19[(jejudata19["업종명"]=="한식 음식점업")
    |(jejudata19["업종명"]=="중식 음식점업")
    |(jejudata19["업종명"]=="일식 음식점업")
    |(jejudata19["업종명"]=="서양식 음식점업")
    |(jejudata19["업종명"]=="비알콜 음료점업")
    |(jejudata19["업종명"]=="기타 외국식 음식점")
    |(jejudata19["업종명"]=="피자, 햄버거, 샌드위치 및 유사 음식점업")]

food_group=food.groupby(["읍면동명"],as_index=False).sum()
food_group["업종명"]="음식"
food_group=food_group.join(food_count)
food_group["점포당 매출액"]=food_group["이용금액"]/food_group["점포 수"]
food_group
```

```python
some=jejudata19[(jejudata19['업종명']=="과실 및 채소 소매업")
    |(jejudata19["업종명"]=="관광 민예품 및 선물용품 소매업")
   |(jejudata19["업종명"]=="수산물 소매업")
    |(jejudata19["업종명"]=="육류 소매업")
    |(jejudata19["업종명"]=="빵 및 과자류 소매업")]
            
some_group=some.groupby(["읍면동명"],as_index=False).sum()
#some_group = pd.merge(some_group, total, how='left', on='읍면동명')
some_group["업종명"]="소매업"
some_group=some_group.join(some_count)
some_group["점포당 매출액"]=some_group["이용금액"]/some_group["점포 수"]
some_group
```

```python
sleep=jejudata19[(jejudata19["업종명"]=="호텔업")
                |(jejudata19["업종명"]=="휴양콘도")
                |(jejudata19["업종명"]=="여관업")]

sleep_group=sleep.groupby(["읍면동명"],as_index=False).sum()
sleep_group["업종명"]="숙박업"
sleep_group=sleep_group.join(sleep_count)
sleep_group["점포당 매출액"]=sleep_group["이용금액"]/sleep_group["점포 수"]
sleep_group
```

```python
activity=jejudata19[(jejudata19["업종명"]=="스포츠 및 레크레이션 용품 임대업")
                    |(jejudata19["업종명"]=="기타 수상오락 서비스업")]

activity_group=activity.groupby(["읍면동명"],as_index=False).sum()
activity_group["업종명"]="문화레저"
activity_group
```

```python
df_ = pd.concat([some_group,sleep_group,food_group, activity_group])
#df_ = df_.sort_values(by=['읍면동명'], axis=0)
df_ = df_.sort_values(by=['업종명','읍면동명'])
df_
```
